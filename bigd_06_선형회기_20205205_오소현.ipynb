```r
# marketing 데이터에서 newspaper outlier 제거
newspaper_upperfence <- summary(marketing$newspaper)[5] + IQR(marketing$newspaper, na.rm=T)*1.5

marketing$newspaper <- ifelse(marketing$newspaper > newspaper_upperfence, NA, newspaper_upperfence)

marketing <- na.omit(marketing)
```

# 선형회귀분석

#8번문제

```r
# setdiff와 sample_frac을 이용한 training/test set생성
print("train_set")
train_set <- sample_frac(marketing, 0.7, replace=F) %>% as_tibble() #70%
train_set %>% show()
```

#9번문제

```r
#setdiff와 sample_frac을 이용한 training/test set생성
print("test_set")
test_set <- setdiff(marketing, train_set) %>% as_tibble() #setdiff() 차집합
test_set %>% show()
```

#10번문제

```r
#youtube(독립변수 or 설명변수)에 따른 sales(종속변수 or 반응변수)의 관계
print("sy_model = sales - youtube")
sy_model <- lm(formula = sales ~ youtube, data = train_set)
summary(sy_model)
```

# 11번문제

- 회귀식: sales=8.121641+0.050788*youtube
- p-value가 유의 수준 보다 작으므로 모델을 채택한다.
- Adjusted R-squared가 0.6067로 61%의 설명력을 갖는다.

# 12번

```r
# newspaper(독립변수 or 설명변수)에 따른 sales(종속변수 or 반응변수)의 관계
print("sn_model = sales ~ newspaper")
sn_model <- lm(formula = sales ~ newspaper, data = train_set)
summary(sn_model)
```

# 13번 문제
